# Latitude and Longitude Finder 

```python
import pandas as pd
import geopandas as gpd
import geopy
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
import matplotlib.pyplot as plt
import folium
from folium.plugins import FastMarkerCluster
import certifi
import ssl
```

```python
ctx = ssl.create_default_context(cafile=certifi.where())
geopy.geocoders.options.default_ssl_context = ctx
```

```python
locator = Nominatim(user_agent="myGeocoder")
location = locator.geocode("Ariana")
```

```python
location.point.latitude
```

```python
def find_latitude_and_longitude(Place):
    location = locator.geocode(Place)
    longitude = location.point.longitude
    latitude = location.point.latitude
    return [latitude,longitude]
```

```python
find_latitude_and_longitude("Tunis")
```

## Load Data

```python
df = pd.read_csv('evolution-nombre-maison-culture.csv')
```

```python
df.head()
```

```python
df = df[["gov_name_f","2016"]]
```

```python
df.head()
```

# Create Map

```python
import folium
tunis_location = find_latitude_and_longitude("Tunis")
tunisia_map = folium.Map(
    location = tunis_location,
    tiles = 'Stamen Terrain',
    zoom_start = 6.49
)
```

```python
tooltip = 'Click!'

for index, row in df.iterrows():
    loc = find_latitude_and_longitude(row["gov_name_f"])
    folium.Marker(loc , popup = row['2016'],  tooltip=tooltip ).add_to(tunisia_map)
```
